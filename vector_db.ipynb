```python
# !pip install pymupdf
# !pip install pdfplumber
# from langchain_community.document_loaders import PyMuPDFLoader
from langchain_community.document_loaders import PDFPlumberLoader
from typing import List, Dict
import os


# [2] PDF 문서 로드 함수
def load_company_pdfs(pdf_directory: str = "data") -> List[Dict]:
    company_docs = []
    total_companies = 0
    
    print("=== PDF 문서 로드 시작 ===")
    
    for filename in os.listdir(pdf_directory):
        if filename.endswith('.pdf'):
            # 파일명에서 기업 정보 추출
            # 예: 기업분석_보고서_국문_뤼튼테크놀로지스_202504300900.pdf
            company_info = filename.split('_')
            company_name = company_info[3]  # 기업명 추출
            report_date = company_info[4].replace('.pdf', '')  # 날짜 추출
            
            file_path = os.path.join(pdf_directory, filename)
            
            try:
                print(f"\n로딩 중: {company_name}")
                loader = PDFPlumberLoader(file_path)
                docs = loader.load()
                
                # 각 페이지에 기업 정보 메타데이터 추가
                for doc in docs:
                    doc.metadata.update({
                        'company_name': company_name,
                        'report_date': report_date,
                        'source_file': filename
                    })
                
                company_docs.extend(docs)
                total_companies += 1
                print(f"완료: {company_name} ({len(docs)} 페이지)")
                
            except Exception as e:
                print(f"에러 발생 ({filename}): {str(e)}")
    
    print(f"\n=== 로드 완료 ===")
    print(f"총 처리된 기업 수: {total_companies}")
    print(f"총 페이지 수: {len(company_docs)}")
    
    return company_docs
# [3] 실행
pdf_directory = "data"
docs = load_company_pdfs(pdf_directory)

# [4] 로드된 문서 확인
def check_loaded_docs(docs):
    print("\n=== 로드된 문서 확인 ===")
    
    # 기업별 페이지 수 집계
    company_pages = {}
    for doc in docs:
        company_name = doc.metadata['company_name']
        company_pages[company_name] = company_pages.get(company_name, 0) + 1
    
    print("\n기업별 페이지 수:")
    for company, pages in company_pages.items():
        print(f"{company}: {pages}페이지")
    
    # 샘플 메타데이터 확인
    print("\n첫 번째 문서의 메타데이터:")
    print(docs[0].metadata)

# 로드된 문서 확인 실행
check_loaded_docs(docs)

# [1] PDF 로드 및 메타데이터 추가
# file_path = "data/기업분석_보고서_국문_뤼튼테크놀로지스_202504300900.pdf"
# company_info = file_path.split('_')
# company_name = company_info[3]
# report_date = company_info[4].replace('.pdf', '')

# loader = PDFPlumberLoader(file_path)
# docs = loader.load()

# # 각 문서에 기업 메타데이터 추가
# for doc in docs:
#     doc.metadata.update({
#         'company_name': company_name,
#         'report_date': report_date,
#         'source_file': file_path
#     })
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

# 표 헤더 패턴 정의
TABLE_HEADERS = [
    "주요 재무상황 (단위 : 백만원)",
    "재무비율분석",
    "주변지역 동종업체 (응용 소프트웨어 개발 및 공급업)",
    "산업재무 분석 (단위: 백만원, %)",
    "주요 수익성비율",
    "수익성비율 해설",
    "주요 안정성비율",
    "안정성비율 해설",
    "주요 성장성 및 활동성비율",
    "주요 성장성 및 활동성비율 해설",
    "재무상태표(GAAP) (단위: 백만원)",
    "차입금구조(GAAP) (단위: 백만원, %)",
    "손익계산서(GAAP) (단위: 백만원, %)",
    "현금흐름분석(간접법) (단위: 백만원, %)",
    "Watch등급 이력",
    "조기경보내역 (단위: 천원)"
]

# 구분자 생성 (각 헤더를 정규식 패턴으로 변환)
separators = []
for header in TABLE_HEADERS:
    # 괄호와 특수문자를 이스케이프 처리
    escaped_header = header.replace("(", "\(").replace(")", "\)")
    separators.append(escaped_header)

# 기본 구분자 추가
separators.extend(["\n\n", "\n", ".", " ", ""])

# 텍스트 스플리터 생성
text_splitter = RecursiveCharacterTextSplitter(
    separators=separators,
    chunk_size=1500,        # 표 데이터를 고려한 더 큰 청크 크기
    chunk_overlap=200,      # 문맥 유지를 위한 오버랩
    length_function=len,
    is_separator_regex=True # 정규식 패턴 사용
)

# 메타데이터 추가를 위한 함수
def add_table_metadata(text: str) -> dict:
    metadata = {
        "document_type": "financial_report",
        "table_type": "unknown"
    }
    
    for header in TABLE_HEADERS:
        if header in text:
            metadata["table_type"] = header
            break
            
    return metadata

# 문서 분할 및 메타데이터 추가
def split_financial_documents(docs):
    processed_chunks = []
    
    for doc in docs:
        chunks = text_splitter.split_text(doc.page_content)
        
        for chunk in chunks:
            metadata = add_table_metadata(chunk)
            metadata.update(doc.metadata)  # 기존 문서의 메타데이터 유지
            
            processed_chunks.append({
                "content": chunk,
                "metadata": metadata
            })
    
    return processed_chunks
```

### Retriever 구현

```python
# !pip install rank_bm25
from langchain.vectorstores import FAISS
from langchain.retrievers import BM25Retriever, EnsembleRetriever
from typing import List, Any
```

```python

def load_hybrid_retriever(vectordb_path: str, docs: List[Any]):
    # Dense
    # 저장된 벡터 DB 로드
    loaded_vectorstore = FAISS.load_local(
        folder_path=vectordb_path,
        embeddings=embeddings,
        allow_dangerous_deserialization=True  # 안전한 소스에서 로드할 때만 True로 설정
    )
    print("벡터 DB 로드 완료!")
    dense_retriever = loaded_vectorstore.as_retriever()

    # Sparse
    sparse_retriever = BM25Retriever.from_documents(docs)

    # Hybrid (dense + sparse)
    hybrid_retriever = EnsembleRetriever(retrievers=[dense_retriever, sparse_retriever], weights=[0.2, 0.8])
    return hybrid_retriever 
```

```python
def search_company_documents(retriever, query: str, company_name: str):
    """특정 기업의 문서만 검색하는 함수"""
    results = retriever.get_relevant_documents(query)
    # 메타데이터의 company_name으로 필터링
    filtered_results = [
        doc for doc in results 
        if doc.metadata.get('company_name') == company_name
    ]
    return filtered_results

# 테스트 코드 수정
def test_hybrid_retriever(retriever, test_queries: List[str]):
    print("=== 하이브리드 리트리버 테스트 ===\n")
    
    for query in test_queries:
        # 쿼리에서 기업명 추출 (예: "뤼튼테크놀로지스의" -> "뤼튼테크놀로지스")
        company_name = query.split('의')[0]
        print(f"[검색 쿼리] {query}")
        print(f"[검색 기업] {company_name}")
        
        try:
            # 특정 기업 문서만 검색
            results = search_company_documents(retriever, query, company_name)
            
            print(f"\n검색된 문서 수: {len(results)}")
            print("\n상위 2개 문서 내용:")
            
            for i, doc in enumerate(results[:5], 1):
                print(f"\n문서 {i}")
                print(f"기업: {doc.metadata.get('company_name', 'N/A')}")
                print(f"보고서 날짜: {doc.metadata.get('report_date', 'N/A')}")
                print(f"문서 유형: {doc.metadata.get('table_type', 'N/A')}")
                print(f"내용 미리보기: {doc.page_content}")
                print("-" * 50)
                
        except Exception as e:
            print(f"검색 중 오류 발생: {e}")
        print("\n" + "=" * 80 + "\n")

# 테스트용 쿼리 목록 (기존과 동일)
test_queries = [
    "뤼튼테크놀로지스의 재무제표는?",
    # "네오사피엔스의 영업이익률 추이는?",
    # "모두싸인의 현금흐름등급은?",
    # "자비스앤빌런즈의 부채비율은?",
    # "엘리스그룹의 신용등급은?"
]

# 리트리버 로드 및 테스트 실행
try:
    # 하이브리드 리트리버 생성
    hybrid_retriever = load_hybrid_retriever(
        vectordb_path="data/vectordb/financial_bge_ko",
        docs=docs
    )
    
    # 테스트 실행
    test_hybrid_retriever(hybrid_retriever, test_queries)
    
except Exception as e:
    print(f"리트리버 테스트 중 오류 발생: {e}")
```

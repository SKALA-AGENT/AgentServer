## 1. GitHub 분석 테스트

```python
github = github_analysis_tool()
company = "뤼튼테크놀로지스"  # 당근마켓 GitHub 조직명

# 조직 정보 조회
org_info = github.get_organization_info(company)
print("\n=== 조직 정보 ===")
print(org_info)

# 주요 레포지토리 조회
repos = github.get_repositories(company)
print("\n=== 주요 레포지토리 ===")
for repo in repos:
    print(f"\n{repo['name']}:")
    print(f"- Stars: {repo['stars']}")
    print(f"- Forks: {repo['forks']}")
    print(f"- Language: {repo['language']}")
    print(f"- Description: {repo['description']}")

# 기여 통계
stats = github.get_contribution_stats(company)
print("\n=== 기여 통계 ===")
print(f"최근 90일 커밋 수: {stats['recent_commits']}")
print(f"활성 기여자 수: {stats['active_contributors']}")
```

## 2. 특허 검색 테스트

```python
# 특허 분석기 인스턴스 생성
patent = patent_search_tool()

# 분석 대상 기업
company = "뤼튼테크놀로지스"

# 특허 분석 실행
result = patent.analyze_patents(company)

# 최근 특허 출력
print("\n=== 최근 특허 ===")
for p in result["patents"]:
    print(f"\n제목: {p['title']}")
    print(
        f"번호: {', '.join(p['patent_numbers']) if p['patent_numbers'] else '정보 없음'}"
    )
    print(f"날짜: {p['published_date']}")
    print(f"요약: {p['summary']}")

# 통계 정보 출력
print("\n=== 특허 통계 ===")
print(f"총 특허 수: {result['patent_count']}")

print("\n연도별 특허 수:")
for year, count in result["year_trend"]:
    print(f"{year}: {count}건")

print("\n상위 기술 분야:")
for field, count in result["top_tech_fields"]:
    print(f"{field}: {count}건")

print("\n기술 키워드 목록:")
print(", ".join(result["tech_keywords"]))
```

## 3. 기술 뉴스 분석 테스트

```python
"""AI 스타트업의 기술력 및 혁신성 분석 사용 예시"""

# 분석 실행
analyzer = tech_news_tool()
company = "필라이즈"

try:
    # 기술력 및 혁신성 분석 실행
    tech_innovation_data = analyzer.analyze_technology_innovation(company)

    # 핵심 기술 키워드 출력
    print(f"\n=== {company}의 핵심 기술 키워드 ===")
    if tech_innovation_data["key_technologies"]:
        for tech, count in tech_innovation_data["key_technologies"]:
            print(f"{tech}: {count}회 언급")
    else:
        print("핵심 기술 키워드를 찾을 수 없습니다.")

    # 기술 혁신 관련 뉴스 출력
    print(f"\n=== {company}의 기술 혁신 관련 뉴스 ===")
    if tech_innovation_data["tech_innovations"]:
        for article in tech_innovation_data["tech_innovations"]:
            print(f"제목: {article['title']}")
            if article.get("published_date"):
                print(f"날짜: {article['published_date']}")
            print(f"URL: {article['url']}")
            print(f"내용: {article['content']}")
            print(f"요약: {article['summary']}")
            print("-" * 50)
    else:
        print("기술 혁신 관련 뉴스를 찾을 수 없습니다.")


except Exception as e:
    print(f"오류 발생: {e}")
    import traceback

    traceback.print_exc()
```
